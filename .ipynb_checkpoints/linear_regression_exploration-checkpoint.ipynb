## Defining Necessary Functions

First, we need to define the primary function we are going to use: linear regression function which is
```
y = m(x) + c

y = y-point
x = x-point
m = gradient
c = y-intercept
```
Basic calculus ezpz

````python
def find_y(x, m, c):
    return (m * x) + c
````

Next, we need to define *cost/loss/error* function. This is the one we are trying to minimize. The intuition is that we are finding certain values of **m** and **c** so that our **cost function** is as close as possible to 0. We will use **mean square error** as our cost function
```
MSE = (1/n) * sum( (Y - y)^2 )

MSE = Mean Square Error
n = number of points
y = y value we get from our equation
Y = the actual Y value
```

````python
def find_error(actual_y, calculated_y):
    return actual_y - calculated_y

def find_square_error(actual_y, calculated_y):
    return (find_error(actual_y, calculated_y))**2

def find_mse(points_dtf, m, c):
    list_of_square_error = points_dtf.apply(lambda x: find_square_error( # This is to find the square error of each row
        find_y(x[0], m, c), x[1]), axis=1) # Use axis=1 so that both columns are captured
    sum_of_square_error = list_of_square_error.sum()
    return sum_of_square_error / list_of_square_error.size
    
````

Let's try the default argument where m and c is 0

````python
find_mse(data, 0, 0)
````

As we want the value of MSE to be as close to 0 as possible, clearly 5000+ isn't really a good number. We need another magic function called gradient descent to solve this

## Gradient Descent

The intuition of this function is to get a certain new value based on a function whose input is the current value. It seems like a mouthful explanation so let's just see the mathematical expression 
```
new_value = old_value - ( learning_rate * function_with_old_value )

new_value = updated value
old_value = current value
learning_rate = how fast the model learns (Gonna visualize this) between 0 to 1
function_with_old_value = resulting value from a function whose input is the current value
```

Now you may be wondering what is the function we are going ot use in this case. Well, We will use partial derivatives of m and c. Why? You can read the intuition more from the Andrew Ng's course or the summary I shared at the start of the page.
```
partial_derivative_of_m = (-1/2n) * sum( x * ( Y - y ) )

n = number of points
x = x value
Y = actual Y value
y = calculated y value

partial_derivative_of_c = (-1/2n) * sum( Y - y )

n = number of points
Y = actual Y value
y = calculated y value

```

This includes partial derivatives too. Read up more on this if you have any questions. Again, the focus here is to visualise the effect if various hyperparameter and not basic ML knowledge

````python
def find_partial_derivative_m(points_dtf, m, c):
    list_of_derivatives = points_dtf.apply(lambda x: x[0] * find_error(x[1], find_y(x[0], m, c)), axis=1)
    return (-1 * list_of_derivatives.sum()) / (2*list_of_derivatives.size)

def find_partial_derivative_c(points_dtf, m, c):
    list_of_derivatives = points_dtf.apply(lambda x: find_error(x[1], find_y(x[0], m, c)) , axis=1)
    return (-1 * list_of_derivatives.sum()) / (2*list_of_derivatives.size)
    
def find_new_c(c, partial_c, learning_rate):
    return c - (learning_rate * partial_c)

def find_new_m(m, partial_m, learning_rate):
    return m - (learning_rate * partial_m)

def find_gradient(points_dtf, m, c, learning_rate):
    partial_m = find_partial_derivative_m(points_dtf, m, c)
    partial_c = find_partial_derivative_c(points_dtf, m, c)
    
    c = find_new_c(c, partial_c, learning_rate)
    m = find_new_m(m, partial_m, learning_rate)
    
    return (m, c)
````

Now we define the main function to run gradient descent

````python
def gradient_descent(points_dtf, initial_m, initial_c, learning_rate, iteration):
    c = initial_c
    m = initial_m
    for x in range(0, iteration):
        m, c = find_gradient(points_dtf, m, c, learning_rate)
    return m, c
````

Let's see the new **m** and **c** as well as the updated **MSE**

````python
m, c =gradient_descent(data, 0, 0, 0.0001, 1000)
````

````python
# New MSE
find_mse(data, m, c)
````

````python
print(m, c)
````

````python
# Visualise our lines
predicted_value = pd.DataFrame()
predicted_value['X'] = data['X']
predicted_value['Y'] = data['X'].apply(lambda x : find_y(x, m, c))

predicted_value.head()
````

````python
# Setting up seaborn superimposed charts
fig, ax = sns.plt.subplots()
sns.regplot(x='X', y='Y', data=data, ax=ax).figure.set_size_inches(20,10)
sns.regplot(x='X', y='Y', data=predicted_value, ax=ax)
````

Not bad, we decrease our MSE to 112 (from 5000+). Our model seems to predict the m and c well

# Part 2

## How m and c changes overtime?

We know that m and c changes overtime but how so?

First we need to store how the value changes overtime and plot it

````python
list_m = []
list_c = []

def find_gradient_v2(points_dtf, m, c, learning_rate):
    partial_m = find_partial_derivative_m(points_dtf, m, c)
    partial_c = find_partial_derivative_c(points_dtf, m, c)
    global list_m
    global list_c
    c = find_new_c(c, partial_c, learning_rate)
    m = find_new_m(m, partial_m, learning_rate)
    list_m.append(m)
    list_c.append(c)
    return (m, c)

def gradient_descent(points_dtf, initial_m, initial_c, learning_rate, iteration):
    c = initial_c
    m = initial_m
    for x in range(0, iteration):
        m, c = find_gradient_v2(points_dtf, m, c, learning_rate)
    return m, c
````

````python
m, c =gradient_descent(data, 0, 0, 0.0001, 1000)
````

````python
x_axis = [x for x in range(0, len(list_m))]

data_m = pd.DataFrame()
data_m['x'] = x_axis
data_m['m'] = list_m

data_c = pd.DataFrame()
data_c['x'] = x_axis
data_c['c'] = list_c
````

````python
sns.pointplot(x='x', y='m', data=data_m).figure.set_size_inches(20,10)
````

````python
sns.pointplot(x='x', y='c', data=data_c).figure.set_size_inches(20,10)
````

Seems like m reaches a point after certain time while c still increases. Shall we see if c actually converges into a value?

````python
list_m = []
list_c = []

m, c =gradient_descent(data, 0, 0, 0.0001, 10000)
````

````python
x_axis = [x for x in range(0, len(list_m))]

data_m = pd.DataFrame()
data_m['x'] = x_axis
data_m['m'] = list_m

data_c = pd.DataFrame()
data_c['x'] = x_axis
data_c['c'] = list_c
````

````python
sns.pointplot(x='x', y='m', data=data_m).figure.set_size_inches(20,10)
````

````python
sns.pointplot(x='x', y='c', data=data_c).figure.set_size_inches(20,10)
````

Seems like c still increases. However, we can safely ignore this as the increase isn't really significant (0.1 after 9000 more iterations; x and y value is much larger than this)

## Conclusion

We can see that we can't set learning rate too small (it will converge very slowly) or too large (it will keep swinging and never converge). We also can see that number iteration will be insignificant when the value already converges. 

**Recommendation: We should define a gradient function that also see the value differences between the old value and the new one. If the value is within certain threshold, we should stop the iteration as thhe value won't likely to change much!**
